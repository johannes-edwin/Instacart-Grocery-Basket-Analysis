# Price 



```python
df_active_customers.shape
```

```python
df_active_customers.columns
```

```python
# define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return np.nan
```

```python
# Apply the function
df_active_customers['prices_range'] = df_active_customers.apply(price_label, axis=1)
```

```python
df_active_customers['prices_range'].value_counts(dropna=False)
```

```python
bar_price_range = df_active_customers['prices_range'].value_counts().plot.bar(color = ['red', 'yellowgreen', 'yellow'])
                                                                                 
plt.title('Distribution of Price Range Groupings')
plt.xlabel('Price Range Group')
plt.xticks(rotation=45)
plt.ylabel('Count')

plt.tight_layout()
```

```python
# Exporting bar chart

bar_price_range.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_price_range_dist.png'), bbox_inches='tight')
```

```python
# Segmentation
Age = {
    'Young (18-30)': '18.5%',      # < 33 tahun
    'Adult (31-50)': '35%',      # 33-49 tahun  
    'Middle-age (51-65)': '35%',  # 50-65 tahun
    'Senior (65+)': '11.5%'       # > 65 tahun
}
```

```python
# Recommended Strategy:
# Product: Focus on the needs of families and middle-aged adults
# Promotion: Content relevant to the 40-60 age group
# Channel: Mix of digital and traditional (due to the average age of 49.5)
```

```python
# Conclusion:
# Instacart customers are predominantly middle-aged (around 50 years old), with a primary focus on the productive age group 30-65.
# The distribution is fairly even, with no extreme dominance of any particular age group.
```

```python
# make def
def categorize_age(age):
    if 18 <= age <= 30:
        return 'Young (18-30)'
    elif 31 <= age <= 50:
        return 'Adult (31-50)'
    elif 51 <= age <= 65:
        return 'Middle Age (51-65)'
    elif age > 65:
        return 'Senior (65+)'
    else:
        return 'Other'  # Untuk usia < 18 jika ada

# Apply function
df_active_customers['age_category'] = df_active_customers['age'].apply(categorize_age)

# Analisis
print("Distribution age category:")
category_dist = df_active_customers['age_category'].value_counts()

for category in ['Young (18-30)', 'Adult (31-50)', 'Middle Age (51-65)', 'Senior (65+)']:
    if category in category_dist:
        count = category_dist[category]
        pct = (count / len(df_active_customers)) * 100
        print(f"{category}: {count:,} ({pct:.1f}%)")

# Cek 'Other' if any
if 'Other' in category_dist:
    print(f"\nOther: {category_dist['Other']:,} data")
```

```python
# Visualizing distribution of customers by age groups.  

# Making a bar chart.

bar_cust_age = df_active_customers['age_category'].value_counts().plot(kind = 'bar', color = ['blue', 'green', 'yellow','red'])

plt.title('Customer Distribution by Age Group')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.ylabel('Count of Customers (in millions)')
plt.tight_layout()
```
