```python
df_merged_large.shape
```

```python
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', how = 'inner')
```

```python
df_merged_large = df_ords.merge(
    df_ords_prior,
    on='order_id',
    how='inner',
    indicator=True
)

df_merged_large['_merge'].value_counts()
```

```python
# ---- Right ---- contoh saja jangan execute
```

```python
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', how = 'right')
```

```python
# --- left --- jangan execute
```

```python
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', how = 'left')
```
