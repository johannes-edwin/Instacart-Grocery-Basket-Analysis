```python
# Someone from the data engineers team in Instacart thinks they’ve spotted something strange about the customer with a "user_id" of “1.” 
# Extract all the information you can about this user.
```

```python
# Filter all data for user_id = 1
user1_orders = df_ords[df_ords['user_id'] == 1]
```

```python
# Preview the results
print(user1_orders)
```

```python
# Summary statistics
print(user1_orders.describe())
```

```python
# Check ordering pattern
print("\nOrders by day of week:")
print(user1_orders['orders_day_of_week'].value_counts())
```

```python
print("\nOrders by hour of day:")
print(user1_orders['order_hour_of_day'].value_counts())
```

```python
# This helps you see:

# Whether they order at unusual hours (e.g., 3 AM)

# If they place orders every day or with unusual gaps

# If any value looks inconsistent (e.g., negative days or skipped order numbers)
```

```python
# Pattern	                                                      Possible Interpretation
# Very small number of orders (e.g., only 4 or 5)	              This user might be a test account or a new customer.
# Identical order_hour_of_day or day_of_week across all rows	  Could indicate automated or test behavior (e.g., always ordering at 7 a.m. Monday).
# Missing or negative days_since_prior_order	                  Data entry issue or corruption.
# Very short intervals between orders (0 or 1 day repeatedly)	  Possibly an internal testing account, not a normal shopper.
# Unusually high order_number (e.g., >100)	                      Account used for system testing or data generation, not a real customer.
```
