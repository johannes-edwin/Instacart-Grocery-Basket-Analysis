```python
for col in df_custs.columns.tolist():
  weird = (df_custs[[col]].applymap(type) != df_custs[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_custs[weird]) > 0:
    print (col)
```

```python
# Check for unique value and type of data in first_name
print("Sample data first_name:")
print(df_custs['first_name'].head(10))

print("\nType unique data in first_name:")
print(df_custs['first_name'].apply(type).value_counts())

# Check row that has trouble
weird_mask = (df_custs[['first_name']].map(type) != df_custs[['first_name']].iloc[0].apply(type)).any(axis=1)
print(f"\nRow with error: {df_custs[weird_mask]['first_name'].tolist()}")
```

```python
# alternate way cleaner
for col in df_custs.columns:
    unique_types = df_custs[col].apply(type).nunique()
    if unique_types > 1:
        print(f"Column {col} has {unique_types} different data type")
        print(df_custs[col].apply(type).value_counts())
```

```python
# Repair troble data
# Conversion all to string
df_custs['first_name'] = df_custs['first_name'].astype(str)

# or handle if any missing values 
print("Missing values di first_name:", df_custs['first_name'].isnull().sum())

# If any NaN that has type= float,then convert to string
df_custs['first_name'] = df_custs['first_name'].fillna('Unknown')
```
